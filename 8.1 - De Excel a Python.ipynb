## Operaciones en Pandas

Vamos creando una columan nueva de ganancias que será igual a precio menos costo de precios y costos, hay que tener en cuenta de que 
los tipos de datos que se exportan desde excel son tipo string y no tipo int para poder hacer las operaciones entonces tener cuidado
con eso 

```python
## forma de cambiar los tipos de datos en excel, en este caso de string a excel 
preciosdf["Precio"] = preciosdf["Precio"].str.replace("$","").astype("float")
```

```python
preciosdf.head(5)
```

```python
preciosdf["Costo"] = preciosdf["Costo"].str.replace("$","").astype("float")
```

```python
## forma para hacer una operacion entre dos columnas !!

preciosdf["Ganancias"] = preciosdf["Precio"] - preciosdf["Costo"]
preciosdf.head(5)
```

Forma de eliminar cualquier columna dentro de pandas, metodo .drop(nombre de la columna) 

```python
restdf = restdf.drop("Hora de Cobro", axis=1)
```

```python
restdf
```

Arma un nuevo dataframe con todos los platillos de más de 100 pesos de precio (usando filtros)

```python
## modo de filtrar datos 

precioscarosdf = preciosdf[preciosdf["Precio"]>100]
precioscarosdf
```

Arma una nueva columna que clasifique platillos en alto/bajo margen, según si su ganancia es mayor a 100 pesos

```python
## ingresar codigo condicional dependiendo, en este caso si se cumple alguna condicion, metodo numpy . where
preciosdf["Margen"]= np.where(preciosdf["Ganancias"]>100,"Alto","Bajo")
preciosdf
```

Cambiale el nombre a la columna de Atendió por Mesero en el df de Restaurant

```python
restdf=restdf.rename(columns={"Atendió":"Mesero"})
restdf
```

## Tablas Dinámicas

Pandas también nos permite analizar datos mediante tablas dinámicas

Vamos creando una tabla dinámica que nos muestre las ventas por categoría

```python
ventaxcat = pd.pivot_table(restpreciodf, values="Precio" , index="Categoria", aggfunc=np.sum)
ventaxcat
```

Ahora una tabla dinámica que nos muestre las ventas por categoría y por tipo en la columna

```python
ventxcatxtipo=pd.pivot_table(restpreciodf, values="Precio", index="Categoria" , columns="Tipo_x",aggfunc=np.sum)
ventxcatxtipo
```

Crea una tabla dináica que muestre las ventas por tipo y categoría (filas), y por mesero (columnas)

```python
tabladin2 = pd.pivot_table(restpreciodf, values = "Precio" , index = ["Tipo_x","Categoria"], columns ="Mesero", aggfunc=np.sum)
tabladin2
```

Crea una tabla dinámica que muestre las ventas promedio por mesero

```python
venprommesero=pd.pivot_table(restpreciodf,values="Precio",index="Mesero")
venprommesero
```
