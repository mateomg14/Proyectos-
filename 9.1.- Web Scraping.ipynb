## Extracción de Datos

Guarda el código HTML de la página en un objeto Contenido

```python
### trae todo el codigo HTML 
contenido = driver.page_source
```

Conviertelo a BeautifulSoup

```python
## lo convierte en codigo HTML para poder leerlo
soup = BeautifulSoup(contenido)
```

```python
##Para filtrar los datps de HTML
soup.find_all("a")
```

Arma un for en el que vayas agregando a nuestras listas los datos necesarios con a.find, usa . append para ir agregando los datos a las listas

```python
for tarjeta in soup.find_all("div",attrs={"class":"sc-i1odl-2 gIHCpf"}):
    area= tarjeta.find("div",attrs={"class":"sc-1uhtbxc-0 cAEjmC"})
    m2sucio.append(area.int)
    
soup.find_all("div",attrs={"class":"sc-i1odl-2 gIHCpf"})
    
        
```

```python

### Donde se van a almacenar los valores
m2_s = []
precio_s = []
ubicacion_s = []
ubicacion_s_2 =[]
descripcion = []
descripcion_corta = []

## En donde se encuentran los valores
atributos = {'class' : 'sc-i1odl-3 VSxgr'}
atributos_area = {'data-qa': 'POSTING_CARD_FEATURES'}
atributos_precio = {'data-qa': 'POSTING_CARD_PRICE' }
atributos_ubicacion = {'data-qa': 'POSTING_CARD_LOCATION'}
atributos_ubicacion_2 = {'class': 'sc-ge2uzh-0 bzGYzE'}
atributos_descripcion = {'data-qa': 'POSTING_CARD_DESCRIPTION'}
atributos_descripcion_corta = {'class': 'sc-i1odl-10 fzzQwx'}

## La forma en el cual se extraen los valores
for tarjeta in soup.find_all('div', attrs = atributos):
    area = tarjeta.find('div', atributos_area)
    m2_s.append(area.text)
    precio = tarjeta.find('div', atributos_precio)
    precio_s.append(precio.text)
    ubicacion = tarjeta.find('div', atributos_ubicacion)
    ubicacion_s.append(ubicacion.text)
    ubicacion_2 = tarjeta.find('div', atributos_ubicacion_2)
    ubicacion_s_2.append(ubicacion_2.text)
    description = tarjeta.find('div', atributos_descripcion)
    descripcion.append(description.text)
    descripcion_corta_ = tarjeta.find('h2', atributos_descripcion_corta)
    descripcion_corta.append(descripcion_corta_.text)

## La forma en que se filtran los valores para traerlos en forma de una tabla     
df =pd.DataFrame({
    'Area':m2_s,
    'Precio':precio_s,
    'Ubicacion':ubicacion_s,
    'Ubicaion_2':ubicacion_s_2,
    'Descripcion':descripcion_corta,
    'Descripción_larga':descripcion
                })
```
